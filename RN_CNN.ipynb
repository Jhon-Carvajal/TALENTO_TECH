```python
import cv2
import matplotlib.pyplot as plt
from google.colab.patches import cv2_imshow
import numpy as np

image = cv2.imread('1.jpg')

kernel = np.array([[0, -1, 0],
                   [-1, 5, -1],
                   [0, -1, 0]])

convolved_images = cv2.filter2D(image, -1, kernel)
plt.figure(figsize=(18,18))
#plt.imshow(image)
plt.imshow(convolved_images)
plt.axis('off')
plt.show()
#cv2_imshow('Imagen Original', image)
#cv2_imshow('Convolved Imagen', convolved_images)
#cv2.waitKey(0)
#cv2.destroyAllWindows()
```
